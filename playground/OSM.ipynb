```python
import osmnx as ox
import matplotlib.pyplot as plt
import networkx as nx
from IPython.display import IFrame
```

```python
ox.config(log_console=True, use_cache=True)
```

```python
G = ox.graph_from_place('New York City, New York, USA', network_type='drive')
```

```python
ox.save_graphml(G, filepath='data/NYC_drive.osm')
```

```python
G=ox.load_graphml(filepath='data/NYC_drive.osm')
G
```

```python
fig, ax = ox.plot_graph(G)
```

```python
origin_node = list(G.nodes())[0]
destination_node = list(G.nodes())[100]
route = nx.shortest_path(G, origin_node, destination_node)
route_d = nx.dijkstra_path(G, origin_node, destination_node)
```

```python
len(route)
```

```python
len(route_d)
```

```python
# plot the route with folium
route_map_length = ox.plot_route_folium(G, route)
```

```python
filepath = 'data/route_length.html'
route_map_length.save(filepath)
IFrame(filepath, width=600, height=500)
```
