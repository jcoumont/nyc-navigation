```python
def get_route_detail(Graph, route_detail):
    node_pairs = zip(route_detail[:-1], route_detail[1:])
    uvk = ((u, v, min(Graph[u][v], key=lambda k: G[u][v][k]["length"])) for u, v in node_pairs)
    gdf_edges = ox.graph_to_gdfs(Graph.subgraph(route_detail), nodes=False).loc[uvk]
    return gdf_edges
```
