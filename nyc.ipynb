```python
# importing modules and librairies

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import folium
from folium import plugins
from folium.plugins import HeatMap
import calendar
```

```python
#loading the datset
df = pd.read_csv('data/data_10000_out_final.csv')
```

```python
# set display options

pd.options.display.max_columns= 100
pd.options.display.max_rows= 100
```

```python
#print the dataset

df.sample(5)
```

```python
heat_df = df[['latitude', 'longitude']]
heat_df = df.dropna(axis=0, subset=['latitude','longitude'])
heat_data = [[row['latitude'],row['longitude']] for index, row in heat_df.iterrows()]
```

```python
# Plot it on the map
traffic_map = folium.Map(location=heat_data[0], zoom_start = 13) 
HeatMap(heat_data).add_to(traffic_map)
traffic_map.save('traffic_map.html')
```

```python
a = fd[['year', 'month', 'day']].astype(str)
fd['date'] = pd.to_datetime(a.agg('-'.join, axis=1), format='%Y-%m-%d')
```

```python
b = fd[['hour', 'minute']].astype(str)
fd['time'] = pd.to_datetime(b.agg('-'.join, axis=1), format='%H-%M')
```

```python
# to review for what i put it
# fd['time'] = [d.time() for d in fd['time']]
```

```python
# Number of accident per day of the week
accidents_day = fd.groupby(fd['date'].dt.dayofweek).count().date

# Replace the day integers by day names.
accidents_day.index=[calendar.day_name[x] for x in range(0,7)]

# plot accidents per day
accidents_day.plot(kind='bar',figsize=(12,7), color='magenta', alpha=0.5)

# title and x,y labels
plt.title('Accidents in NYC', fontsize=20)
plt.xlabel('Day of the week',fontsize=16)
plt.ylabel('Number of accidents',fontsize=16);
```
